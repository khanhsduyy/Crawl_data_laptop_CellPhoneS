```python
# Function get all link laptop old on page
def GetUrl():
    pageSource = BeautifulSoup(driver.page_source)
    infoTag = pageSource.find_all('a', 'product__link button__link')
    urls = []
    for info in infoTag:
        url = info.get('href')
        urlLaptop = 'https://cellphones.com.vn' + url
        if urlLaptop not in urls:
            urls.append(urlLaptop)
    return urls
```

```python
# Get all url
urls = GetUrl()
```

```python
# Function get Price Laptop
def GetPriceOld(pageSource):
    priceLaptop = None
    try:
        priceFieldTagP = pageSource.find('p', 'tpt---price')
        textPriceLaptop = priceFieldTagP.get_text().strip()
        priceLaptop = textPriceLaptop
    except:
        return priceLaptop
    return priceLaptop
```

```python
# Function get Price Laptop
def GetPriceNew(pageSource):
    priceLaptop = None
    try:
        priceFieldTagP = pageSource.find('p', 'tpt---sale-price')
        textPriceLaptop = priceFieldTagP.contents[0].strip()
        priceLaptop = textPriceLaptop
    except:
        return priceLaptop
    return priceLaptop
```

### Get name laptop

```python
# Function get name all laptop
def GetName(pageSource):
    nameLaptop = None
    try:
        nameField = pageSource.find('div', 'box-product-name')
        textName = nameField.get_text().strip()
        nameLaptop = textName
    except:
        return nameLaptop
    return nameLaptop

```

###  Get warranty laptop

```python
def GetWarranty(pageSource):
    warrantyLaptop = None
    try:
        infoWarrantyField = pageSource.find_all('div', 'description')
        textWarranty = infoWarrantyField[1].get_text().strip()
        warrantyLaptop = textWarranty
    except:
        warrantyLaptop
    return warrantyLaptop
```

### Get info laptop

```python
# Function get info CPU
def GetInfo(pageSource):
    CARD = None
    CPU = None
    RAM_SIZE = None
    RAM_TYPE = None
    HARD_DRIVER = None
    TOUCH_SCREEN = None
    PANEL = None
    SCREEN_SIZE = None
    RESOLUTION = None
    OS = None
    SIZE = None
    WEIGHT = None
    
    try:
        tagLiField = pageSource.find_all('div', 'modal-item-description mx-2')
        for index, element in enumerate(tagLiField):
            tagDivField = tagLiField[index].find_all('div', 'px-3 py-2 is-flex is-align-items-center is-justify-content-space-between')
            for i, ele in enumerate(tagDivField):
                if ele.find('p').get_text().strip() == 'Loại card đồ họa':
                    CARD = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Loại CPU':
                    CPU = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Dung lượng RAM':
                    RAM_SIZE = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Loại RAM':
                    RAM_TYPE = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Ổ cứng':
                    HARD_DRIVER = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Màn hình cảm ứng':
                    TOUCH_SCREEN = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Chất liệu tấm nền':
                    PANEL = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Kích thước màn hình':
                    SCREEN_SIZE = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Độ phân giải màn hình':
                    RESOLUTION = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Hệ điều hành':
                    OS = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Kích thước':
                    SIZE = ele.find('div').get_text().strip()
                if ele.find('p').get_text().strip() == 'Trọng lượng':
                    WEIGHT = ele.find('div').get_text().strip()
    except:
        return CARD, CPU, RAM_SIZE, RAM_TYPE, HARD_DRIVER, TOUCH_SCREEN, PANEL, SCREEN_SIZE, RESOLUTION, OS, SIZE, WEIGHT
    return CARD, CPU, RAM_SIZE, RAM_TYPE, HARD_DRIVER, TOUCH_SCREEN, PANEL, SCREEN_SIZE, RESOLUTION, OS, SIZE, WEIGHT
```

```python
data = []
def main():
    index = 0
    for url in urls:
        driver.get(url)
        sleep(1)
        pageSourceBefore = BeautifulSoup(driver.page_source)
        nameLaptop = GetName(pageSourceBefore)
        priceLaptopOld = GetPriceOld(pageSourceBefore)
        priceLaptopNew = GetPriceNew(pageSourceBefore)
        warrantyLaptop = GetWarranty(pageSourceBefore)
        
        if nameLaptop == None:
            continue
            
        heightPage = 1200
        while True:
            # Load page
            driver.execute_script('window.scrollTo(0, {})'.format(heightPage))
            driver.execute_script('window.scrollTo(0, {})'.format(heightPage - 400))
            driver.execute_script('window.scrollTo(0, {})'.format(heightPage + 600))
            driver.execute_script('window.scrollTo(0, {})'.format(heightPage - 200))
            
            button = 0
            y = 0
            try:
                sleep(1)
                button = driver.find_element(By.XPATH, 
                                             '//*[@id="productDetailV2"]/section/div[4]/div[2]/div/div/button')
                location = button.location
                y = location['y']
            except:
                button = button
            if button != 0:
                driver.execute_script('window.scrollTo(0, {})'.format(y - 500))
                sleep(1)
                button.click()
                break
            
            
        pageSourceAfter = BeautifulSoup(driver.page_source)
        CARD, CPU, RAM_SIZE, RAM_TYPE, HARD_DRIVER, TOUCH_SCREEN, PANEL, SCREEN_SIZE, RESOLUTION, OS, SIZE, WEIGHT = GetInfo(pageSourceAfter)
         
        data.append({'name': nameLaptop, 'priceOld': priceLaptopOld, 'priceNew':priceLaptopNew, 'warranty': warrantyLaptop, 'card': CARD,
                    'cpu': CPU, 'ram_size': RAM_SIZE, 'ram_type': RAM_TYPE, 'hard_driver': HARD_DRIVER, 'touch_screen': TOUCH_SCREEN,
                    'panel': PANEL, 'screen_size': SCREEN_SIZE, 'resolution': RESOLUTION, 'os': OS, 'size': SIZE, 'weight': WEIGHT})
        print('Crawl sucess laptop: ', index)
        index += 1
    driver.quit()
main()
print(len(data))
```
